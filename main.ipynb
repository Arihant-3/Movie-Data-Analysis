```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('movies_data.csv') 
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
sns.histplot(df['averageRating'], bins=20, kde=True, color="blue")
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Split the genres column by commas, creating a list for each movie
df['genres'] = df['genres'].str.split(', ')

# Explode the list so that each genre appears as a separate row
df_exploded = df.explode('genres')

# Group by genre and count the number of occurrences
genre_count = df_exploded['genres'].value_counts()
genre_count
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x=genre_count.index, y=genre_count.values, palette="Spectral", hue=None, legend=False)
plt.title('Distribution of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df["title"].value_counts()
```

```python
years = df["releaseYear"].value_counts().sort_index()
years
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years.index, years.values, marker='.', linestyle='-', color='c')

plt.title('Number of Movies Released Per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')

plt.tight_layout()
plt.show()
```
